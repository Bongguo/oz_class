```python
import random

# ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter + " "
        else:
            display += "_ "
    return display.strip()

def hangman():
    word_to_guess = random.choice(words)
    guessed_letters = []
    attempts_left = 10

    print("Welcome to Hangman!")

    while attempts_left > 0:
        current_display = display_word(word_to_guess, guessed_letters)
        print(f"\nCurrent word: {current_display}")

        user_guess = input("Guess a letter: ").lower()

        if len(user_guess) != 1 or not user_guess.isalpha():
            print("Please enter a single alphabetical character.")
            continue

        if user_guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        guessed_letters.append(user_guess)

        if user_guess in word_to_guess:
            print("Correct! Keep it up!")
        else:
            attempts_left -= 1
            print(f"Wrong guess. Remaining attempts: {attempts_left}")

        if set(guessed_letters) == set(word_to_guess):
            print(f"Congratulations! You guessed the word '{word_to_guess}' with {10 - attempts_left} attempts left.")
            break

    if attempts_left == 0:
        print(f"\nSorry, you're out of attempts. The correct word was '{word_to_guess}'.")

if __name__ == "__main__":
    hangman()
```
